```python
# 1. Bring in the pandas library
import pandas as pd

# 2. Tell pandas to read our CSV file and store it in a variable named 'df' (dataframe)
# Note: If your file has a slightly different name, replace it here!
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# 3. Display the first 5 rows of our data
df.head()
```

```python
# Check the data types and look for missing values
df.info()
```

```python
# 1. Force the TotalCharges column to become numbers
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# 2. Check our work! Let's count how many missing values (NaNs) we have in each column now
df.isnull().sum()
```

```python
# 1. Drop any row that has a missing value (NaN)
df.dropna(inplace=True)

# 2. Check the final size (shape) of our dataset
df.shape
```

```python
# 1. Bring in the Matplotlib library to draw our charts
import matplotlib.pyplot as plt

# 2. Group our data to count how many people churned vs stayed, separated by their contract type
contract_churn = df.groupby(['Contract', 'Churn']).size().unstack()

# 3. Build the visual structure (like setting up a canvas)
contract_churn.plot(kind='bar', color=['#4CAF50', '#F44336'], figsize=(8, 5))

# 4. Add clear labels so the information is instantly readable
plt.title('Churn Patterns by Contract Type')
plt.xlabel('Contract Length')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)

# 5. Render the final chart to the screen
plt.show()
```

```python
# Save our cleaned DataFrame to a new CSV file without the index numbers
df.to_csv('Cleaned_Telco_Data.csv', index=False)

from google.colab import files
files.download('Cleaned_Telco_Data.csv')
```
